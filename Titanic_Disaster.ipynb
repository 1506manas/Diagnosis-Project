```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Data_Files/train.csv")
df_test = pd.read_csv("Data_Files/test.csv")
print(df)
```

```python
df = df.drop(labels = 'Cabin', axis = 1)
```

```python
df.info()
```

# For Train Data

```python
from sklearn.impute import SimpleImputer
num_val_train_df = df.select_dtypes(include = ["int64","float64"]).columns
df[num_val_train_df].isnull().sum()
```

```python
imputer_mean = SimpleImputer(strategy = 'mean')
imputer_mean.fit(df[num_val_train_df])
df[num_val_train_df] = imputer_mean.transform(df[num_val_train_df])
df[num_val_train_df].isnull().sum()
```

```python
df['Embarked'].unique()
```

```python
df1 = df.dropna(axis = 0)
df1
```

```python
df1['Embarked'].replace(['S', 'C', 'Q'], (0,1,2), inplace = True)
df1['Sex'].replace(['male', 'female'], (0,1), inplace = True)
```

```python
df1.info()
```

```python
df1
```

```python
x_train = df1[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked']]
y_train = df1['Survived']
```

# For Test Data

```python
num_val_train_df_test = df_test.select_dtypes(include = ["int64","float64"]).columns
df_test[num_val_train_df_test].isnull().sum()
```

```python
imputer_mean1 = SimpleImputer(strategy = 'mean')
imputer_mean1.fit(df_test[num_val_train_df_test])
df_test[num_val_train_df_test] = imputer_mean1.transform(df_test[num_val_train_df_test])
df_test[num_val_train_df_test].isnull().sum()
```

```python
df_test['Embarked'].replace(['S', 'C', 'Q'], (0,1,2), inplace = True)
df_test['Sex'].replace(['male', 'female'], (0,1), inplace = True)
df_test
```

```python
df_test.info()
```

```python
from sklearn.svm import SVC
classification_linear = SVC(kernel='linear')
classification_linear.fit(x_train, y_train)
```

```python
x_test = df_test[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked']]
```

```python
y_pred = classification_linear.predict(x_test)
y_pred
```

```python
df_test['Survived'] = y_pred
df_test
```

```python
df_test.to_csv('df_test.csv')
```
